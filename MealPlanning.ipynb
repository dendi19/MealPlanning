```python
import pandas as pd
```

```python
df = pd.read_csv('allmeals.csv')
```

```python
df
```

```python
breakfasts = []
lunches = []
dinners = []

for table, meals in df.iterrows():
    if meals[1] == 'Breakfast':
        #print(f'{meals[0]} is breakfast!')
        breakfasts.append(meals[0])
    elif meals[1] == 'Lunch':
        lunches.append(meals[0])
    elif meals[1] =='Dinner':
        dinners.append(meals[0])
```

```python
import random
from datetime import date
import datetime

today = date.today()
```

```python
dotw = ['Meal','Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']
thisweek = pd.DataFrame(columns=dotw)

brekkylist = ['Breakfast']
lunchylist = ['Lunch']
dinnerslist = ['Dinner']

for items in dotw:
    if items == 'Meal':
        pass
    else:
        brekkylist.append(random.choice(breakfasts))
print(brekkylist)
thisweek = thisweek.append({'Meal': brekkylist[0],'Sunday': brekkylist[1], 'Monday': brekkylist[2], 'Tuesday': brekkylist[3], 'Wednesday': brekkylist[4], 'Thursday': brekkylist[5], 'Friday': brekkylist[6], 'Saturday': brekkylist[7]}, ignore_index=True)

for items in dotw:
    if items == 'Meal':
        pass
    else:
        lunchylist.append(random.choice(lunches))
print(lunchylist)
thisweek = thisweek.append({'Meal': lunchylist[0],'Sunday': lunchylist[1], 'Monday': lunchylist[2], 'Tuesday': lunchylist[3], 'Wednesday': lunchylist[4], 'Thursday': lunchylist[5], 'Friday': lunchylist[6], 'Saturday': lunchylist[7]}, ignore_index=True)
 

for items in dotw:
    meal = random.choice(dinners)
    if items == 'Meal':
        pass
    else:
        dinnerslist.append(meal)
print(dinnerslist)
thisweek = thisweek.append({'Meal': dinnerslist[0],'Sunday': dinnerslist[1], 'Monday': dinnerslist[2], 'Tuesday': dinnerslist[3], 'Wednesday': dinnerslist[4], 'Thursday': dinnerslist[5], 'Friday': dinnerslist[6], 'Saturday': dinnerslist[7]}, ignore_index=True)
```
